```python
members = [485, 834, 979, 1359, 1586, 1699, 1918, 2482] # тестовая выборка для отладки исключений
```

```python
from collections import defaultdict
number_errors = 0 # счетчик неуспешно проанализированных пользователей на подписки
number_success = 0 # счетчик успешно проанализированных пользователей на подписки
groups = defaultdict(list) # специальный словарь, где каждому ключу соответсвует list значений
for m_key in members:
    try:
        dictOfSub = vk_api.users.getSubscriptions(user_id = m_key) # выводим подписки для члена сообщества номер m
        listOfSubGr = []
        for key in dictOfSub.keys(): # выполняем цикл по ключам в выводе подписок, так как формат подписок не удовлетворяет нашим требованиям
            if key == 'groups': 
                listOfSubGr = dictOfSub[key] # извлекаем только лист, ключом которого является "groups"
        listOfSubClear = [] # создаем list в котором будут только идентификаторы групп
        for key in listOfSubGr.keys(): 
            if key == 'items':
                listOfSubClear = listOfSubGr[key]
        for l_key in listOfSubClear: # итератор по всем подпискам опрделенного человека
            groups[l_key].append(m_key) # заполнения словаря, где ключ это id группы, а значения это list пользователей
        number_success += 1
    except Exception:
        number_errors += 1
        pass
    
print(len(groups.keys()))
print(number_errors)
print(number_success)
```
